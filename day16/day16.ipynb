```python
class Book:
    def __init__(sl):
        sl.title="Python"
    
    def dis(sl):
        print("sl is printed", sl)
        print("type of self is printed", type(sl))

b1=Book()
print("b1 is printed", b1)
b1.dis()
print("type of b1 is printed", type(b1))

print(Book())
```

```python
class Book:

    def __init__(sl):
        print(id(sl))

b1 = Book()

print(id(b1))

#Both id() values will be identical.
# This proves mathematically that
# sl and b1 are not two different objects.
# They are two references pointing to the exact same object in memory.
```
