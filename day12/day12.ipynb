```python
#CONCATENATION
# Using the + Operator
netflix_shows = ["Stranger Things", "The Crown", "Ozark", "Dark", "Breaking Bad", "Black Mirror", "Mindhunter", "Squid Game"]
new_releases = ["The Witcher", "Lupin"]

all_shows = new_releases + netflix_shows

print(all_shows)
```

```python
# Using the extend() Method
netflix_shows.extend(new_releases)
print(netflix_shows)
```

## Important Differences:
+ (Concatenation): Use this when you want to keep the original lists unchanged and create a third, combined list.

extend(): Use this when you want to update an existing list and save memory by not creating a new one.

Pro-tip for your class:
Be careful not to confuse concatenation with append().

If you do list1.append(list2), you will get a nested list (a list inside a list).

If you do list1 + list2 or list1.extend(list2), you get a flat list where all items are at the same level.

# sorted(list)
This is one of the most useful built-in functions.

What it does: It takes any list and returns a new list with the items in ascending order (smallest to largest).

Important: It does not change the original list; it creates a copy.

# reversed(list)
This is similar to slicing [::-1], but it returns an iterator.

What it does: It lets you loop through a list backwards without actually creating a new list in memory (which is great for performance).

Note: To see the result as a list, you must wrap it in list().

⚠️ remember: sorted() is safer because it doesn't destroy your original data!


#Our sample data

# List Comprehension
### List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.
Without list comprehension you will have to write a for statement with a conditional test inside:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

With list comprehension you can do all that with only one line of code:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```

## The Syntax
####  newlist = [expression for item in iterable if condition == True]

```python
#Only accept items that are not "apple":
newlist = [x for x in fruits if x != "apple"]
print(newlist)

#With no if statement:
newlist = [x for x in fruits]
print(newlist)

#You can use the range() function to create an iterable:
newlist = [x for x in range(10)]
print(newlist)

#Accept only numbers lower than 5:
newlist = [x for x in range(10) if x < 5]
print(newlist)
```

## remove(), pop(), and del

### remove(value): 
Use when you know the name/value of the item you want to delete, but don't know where it is. (Only works on lists, not dictionaries).

### pop(index/key): 
Use when you know the position/key of the item, AND you want to save that item into a variable to use it right after removing it.

### del: 
Use when you know the position/key of the item, and you just want it completely gone without saving it. You can also use it to delete entire variables from your computer's memory!

```python
numbers = [10, 20, 30, 40, 50]

numbers.remove(30)  # List se '30' gayab ho jayega
print(numbers)      
# Output: [10, 20, 40, 50]

popped_item = numbers.pop(1)  # Index 1 par '20' hai, usey nikaal dega
print(numbers)                 # Output: [10, 40, 50]
print(popped_item)            # Output: 20  (jo nikal gaya)

del numbers[0]  # Pehla element '10' direct khatam
print(numbers)  
# Output: [40, 50]
```

### sort() & sorted()

```python
# Original list
numbers = [40, 10, 30, 20]

# Using sorted() function
new_list = sorted(numbers)

print("Original List:", numbers)   # Output: [40, 10, 30, 20] -> Remains untouched!
print("New Sorted List:", new_list) # Output: [10, 20, 30, 40]
```

```python
# Both sort() and sorted() accept a special argument called reverse=True if 
# you want to sort from highest to lowest.

# Descending order with sorted()
scores = [5, 12, 8, 20]
high_to_low = sorted(scores, reverse=True)

print(high_to_low)  
# Output: [20, 12, 8, 5]


# Descending order with sort()
scores.sort(reverse=True)

print(scores)       
# Output: [20, 12, 8, 5]
```
