LISTS COMPREHENSION

```python
#CONCATENATION
# Using the + Operator
netflix_shows = ["Stranger Things", "The Crown", "Ozark", "Dark", "Breaking Bad", "Black Mirror", "Mindhunter", "Squid Game"]
new_releases = ["The Witcher", "Lupin"]

all_shows = new_releases + netflix_shows

print(all_shows)
```

```python
# Using the extend() Method
netflix_shows.extend(new_releases)
print(netflix_shows)
```

## Important Differences:
+ (Concatenation): Use this when you want to keep the original lists unchanged and create a third, combined list.

extend(): Use this when you want to update an existing list and save memory by not creating a new one.

Pro-tip for your class:
Be careful not to confuse concatenation with append().

If you do list1.append(list2), you will get a nested list (a list inside a list).

If you do list1 + list2 or list1.extend(list2), you get a flat list where all items are at the same level.

## AGGREGATE FXNS
# 1. The Standard Aggregates
You used these in your Company Payroll Analytics section:

sum(list): Adds all elements together.

max(list): Returns the largest item.

min(list): Returns the smallest item.

len(list): Returns the count of items. (Note: len() is technically a "counting" function, but it acts as an aggregate because it summarizes the collection).

# 2. The "Average" Trick
Python does not have a built-in avg() function. Instead, we calculate the average by aggregating two different functions together: sum() divided by len().

average = sum(salaries) / len(salaries)


# 3. Advanced Aggregates (The statistics module)

import statistics

data = [45000, 60000, 32000, 85000, 52000]

#Median: The middle value when the list is sorted
print(statistics.median(data)) 

#Mode: The most frequently occurring value
print(statistics.mode(data))

#Stdev: The standard deviation (how spread out the data is)
print(statistics.stdev(data))


# Why call them "Aggregate" functions?
They are called "aggregate" because they gather multiple pieces of data and condense them into one single representative value.

# sorted(list)
This is one of the most useful built-in functions.

What it does: It takes any list and returns a new list with the items in ascending order (smallest to largest).

Important: It does not change the original list; it creates a copy.

# reversed(list)
This is similar to slicing [::-1], but it returns an iterator.

What it does: It lets you loop through a list backwards without actually creating a new list in memory (which is great for performance).

Note: To see the result as a list, you must wrap it in list().

⚠️ remember: sorted() is safer because it doesn't destroy your original data!


#Our sample data
data = [45, 12, 89, 30]

#--- 1. Using sorted() (Non-destructive) ---
#We keep the original list, and create a new sorted one
sorted_data = sorted(data, reverse=True)

print("Original data remains:", data)       # Still [45, 12, 89, 30]
print("New descending list:", sorted_data)  # Becomes [89, 45, 30, 12]

#--- 2. Using list.sort() (In-place/Destructive) ---
#We modify the original list directly
data.sort(reverse=True)

print("\nData after .sort() modification:", data) # Becomes [89, 45, 30, 12]

# any() and all()
These are boolean aggregators. They check the truthiness of items in an iterable.

any(list): Returns True if at least one item in the list is True (or non-zero).

all(list): Returns True only if every single item in the list is True.
eg :is_anyone_high_risk = any([hr > 100 for hr in heart_rates])
print(f"Is there any high risk patient? {is_anyone_high_risk}")

# enumerate()
shows = ["Stranger Things", "The Crown", "Ozark"]
#Using enumerate to get both index and value
#We can start the index at 1 instead of 0 if we want!
for index, show in enumerate(shows, start=1):
    print(f"Show #{index}: {show}")


creating a dictionary from a list using enumerate:
show_dict = {index: show for index, show in enumerate(shows)}
