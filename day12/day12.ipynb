```python
#CONCATENATION
# Using the + Operator
netflix_shows = ["Stranger Things", "The Crown", "Ozark", "Dark", "Breaking Bad", "Black Mirror", "Mindhunter", "Squid Game"]
new_releases = ["The Witcher", "Lupin"]

all_shows = new_releases + netflix_shows

print(all_shows)
```

```python
# Using the extend() Method
netflix_shows.extend(new_releases)
print(netflix_shows)
```

## Important Differences:
+ (Concatenation): Use this when you want to keep the original lists unchanged and create a third, combined list.

extend(): Use this when you want to update an existing list and save memory by not creating a new one.

Pro-tip for your class:
Be careful not to confuse concatenation with append().

If you do list1.append(list2), you will get a nested list (a list inside a list).

If you do list1 + list2 or list1.extend(list2), you get a flat list where all items are at the same level.

# List Comprehension
### List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Based on a list of fruits, you want a new list, containing only the fruits with the letter "a" in the name.
Without list comprehension you will have to write a for statement with a conditional test inside:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
```

With list comprehension you can do all that with only one line of code:

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]
print(newlist)
```

## The Syntax
####  newlist = [expression for item in iterable if condition == True]

```python
#Only accept items that are not "apple":
newlist = [x for x in fruits if x != "apple"]
print(newlist)

#With no if statement:
newlist = [x for x in fruits]
print(newlist)

#You can use the range() function to create an iterable:
newlist = [x for x in range(10)]
print(newlist)

#Accept only numbers lower than 5:
newlist = [x for x in range(10) if x < 5]
print(newlist)
```

## remove(), pop(), and del

### remove(value): 
Use when you know the name/value of the item you want to delete, but don't know where it is. (Only works on lists, not dictionaries).

### pop(index/key): 
Use when you know the position/key of the item, AND you want to save that item into a variable to use it right after removing it.

### del: 
Use when you know the position/key of the item, and you just want it completely gone without saving it. You can also use it to delete entire variables from your computer's memory!

```python
numbers = [10, 20, 30, 40, 50]

numbers.remove(30)  # List se '30' gayab ho jayega
print(numbers)      
# Output: [10, 20, 40, 50]

popped_item = numbers.pop(1)  # Index 1 par '20' hai, usey nikaal dega
print(numbers)                 # Output: [10, 40, 50]
print(popped_item)            # Output: 20  (jo nikal gaya)

del numbers[0]  # Pehla element '10' direct khatam
print(numbers)  
# Output: [40, 50]
```

### sort() & sorted()

```python
# Original list
numbers = [40, 10, 30, 20]

# Using sorted() function
new_list = sorted(numbers)

print("Original List:", numbers)   # Output: [40, 10, 30, 20] -> Remains untouched!
print("New Sorted List:", new_list) # Output: [10, 20, 30, 40]
```

```python
# Both sort() and sorted() accept a special argument called reverse=True if 
# you want to sort from highest to lowest.

# Descending order with sorted()
scores = [5, 12, 8, 20]
high_to_low = sorted(scores, reverse=True)

print(high_to_low)  
# Output: [20, 12, 8, 5]


# Descending order with sort()
scores.sort(reverse=True)

print(scores)       
# Output: [20, 12, 8, 5]
```

Library management system: store books, add books, search issue, return, create backup, sort books and display final library using appropriate list methods.

```python
# --- INITIAL SYSTEM DATA ---
# Main library catalog storing active book titles
library_books = ["The Hobbit", "1984", "To Kill a Mockingbird", "The Great Gatsby"]
# List to track books that are currently issued to students
issued_books = []

print("=== Welcome to the Library Management System ===")
print("Initial Library Catalog:", library_books)

# ---------------------------------------------------------
# 1. ADD BOOKS (Using append)
# ---------------------------------------------------------
print("\n--- 1. Adding New Arrivals ---")
library_books.append("Dune")
library_books.append("Neuromancer")
print("Catalog after adding new books:", library_books)

# ---------------------------------------------------------
# 2. SEARCH & ISSUE BOOK (Using membership checks and remove)
# ---------------------------------------------------------
print("\n--- 2. Issuing a Book ---")
search_title = "1984"

# Check if the book exists in the library
if search_title in library_books:
    print(f"Success: '{search_title}' is available in stock!")
    # Remove from available catalog and add to issued tracking list
    library_books.remove(search_title)
    issued_books.append(search_title)
    print(f"Action: '{search_title}' has been successfully issued.")
else:
    print(f"Error: Sorry, '{search_title}' is currently unavailable or already issued.")

print("Current Available Catalog:", library_books)
print("Currently Issued Books   :", issued_books)

# ---------------------------------------------------------
# 3. RETURN BOOK (Using append)
# ---------------------------------------------------------
print("\n--- 3. Returning an Issued Book ---")
return_title = "1984"

if return_title in issued_books:
    # Remove from issued list and add back to library list
    issued_books.remove(return_title)
    library_books.append(return_title)
    print(f"Action: '{return_title}' successfully received and returned to the shelves.")
else:
    print(f"Error: '{return_title}' does not belong to the issued list.")

# ---------------------------------------------------------
# 4. CREATE BACKUP (Using copy)
# ---------------------------------------------------------
print("\n--- 4. Creating Database Backup ---")
# Creates an independent copy so modifications to one don't corrupt the other
library_backup = library_books.copy()
print("Backup created successfully. Total items backed up:", len(library_backup))

# ---------------------------------------------------------
# 5. SORT BOOKS (Using sort)
# ---------------------------------------------------------
print("\n--- 5. Sorting Catalog Alphabetically ---")
# Permanently sorting the active catalog in-place
library_books.sort()
print("Sorting complete.")

# ---------------------------------------------------------
# 6. DISPLAY FINAL SYSTEM STATE
# ---------------------------------------------------------
print("\n=================================================")
print("             FINAL LIBRARY STATUS                ")
print("=================================================")
print(f"Available Books ({len(library_books)}) : {library_books}")
print(f"Issued Books    ({len(issued_books)}) : {issued_books}")
print(f"Database Backup ({len(library_backup)}) : {library_backup}")
print("=================================================")
```
