```python
n = int(input("Enter number of rows: "))

print("\nPattern 5")
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

```python
# pattern 10
n = int(input("Enter a number: "))

print("\nPattern10")

for i in range(1, n + 1):
    for j in range(n, i - 1, -1):
        print(j, end="")
    print()
```

```python
#a scenario of a hospital whih have few departments in the different building
#  the package of medicined have to be delivered according to the builinf then in each
#  building to each flooe an d in each floor every rooms

b = int(input("Enter number of buildings: "))
f = int(input("Enter number of floors per building: "))
r = int(input("Enter number of rooms per floor: "))

print("\nStarting Medicine Delivery...\n")

for i in range(1, b + 1):
    print(f"--- Arrived at Building {i} ---")
    
    for j in range(1, f + 1):
        for k in range(1, r + 1):
            print(f"Delivered to Building {i} | Floor {j} | Room {k}")
            
    print(f"--- Finished Building {i} ---\n")

print("All medicine packages have been successfully delivered!")
```
