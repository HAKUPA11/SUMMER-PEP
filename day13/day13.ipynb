```python
#### SET OPERATIONS

#### ==============================================================================================
####                                   THE BIG FOUR SET OPERATIONS
#### ==============================================================================================
####  Operation            | Symbol | Method                 | What it does
#### ----------------------|--------|------------------------|-------------------------------------
####  Union                |   |    | .union()               | Dono sets ke saare elements ko mila deta hai (removes duplicates).
####  Intersection         |   &    | .intersection()        | Sirf wahi elements rakhta hai jo dono sets mein common hain.
####  Difference           |   -    | .difference()          | Pehle set ke wo elements rakhta hai jo dusre set mein nahi hain.
####  Symmetric Difference |   ^    | .symmetric_difference()| Wo elements rakhta hai jo kisi ek set mein hain, par dono mein nahi.
#### ==============================================================================================





#### Quick Code Example to test it:
set_a = {1, 2, 3}
set_b = {3, 4, 5}

#### 1. Union Example
print("Union:", set_a | set_b)                      # Output: {1, 2, 3, 4, 5}

#### 2. Intersection Example
print("Intersection:", set_a.intersection(set_b))    # Output: {3}

# Dono sets mein 'Amit' aur 'Priya' common hain
dance_club = {"Rahul", "Anjali", "Amit", "Priya"}
music_club = {"Amit", "Priya", "Rohan", "Siddharth"}

# Dance club minus Music club (Sirf Pure Dancers bachenge)
only_dance = dance_club - music_club
print("Difference (Dance - Music):", only_dance)
# Output: {'Rahul', 'Anjali'}

# Music club minus Dance club (Sirf Pure Musicians bachenge)
only_music = music_club - dance_club
print("Difference (Music - Dance):", only_music)
# Output: {'Rohan', 'Siddharth'}


# Wo log jo ya toh sirf dance mein hain ya sirf music mein (Common wale gayab)
either_but_not_both = dance_club ^ music_club
print("Symmetric Difference:", either_but_not_both)
# Output: {'Rahul', 'Anjali', 'Rohan', 'Siddharth'}
```

#### SET METHODS

```python
# ==============================================================================================
#                                   IMPORTANT SET METHODS
# ==============================================================================================
#  Method     | Syntax               | What it does
# ------------|----------------------|----------------------------------------------------------
#  add        | set.add(item)        | Set mein ek naya element add karta hai.
#  update     | set.update(iterable) | Ek sath kayi saare elements (list ya set) add karta hai.
#  remove     | set.remove(item)     | Element delete karta hai. Agar element na ho toh ERROR deta hai.
#  discard    | set.discard(item)    | SAFE DELETE! Element na bhi ho toh ERROR NAHI deta.
#  pop        | set.pop()            | Kisi bhi RANDOM element ko nikaal kar delete kar deta hai.
#  clear      | set.clear()          | Set ke saare elements ko uda kar usey khaali (empty) kar deta hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

my_set = {"Python", "Java"}

# 1. add() — Ek item add karna
my_set.add("C++")
print("After add:", my_set)          # Output: {'Python', 'Java', 'C++'} (Order random ho sakta hai)


# 2. update() — Ek sath multiple items lana
my_set.update(["JavaScript", "Go"])
print("After update:", my_set)       # Output: {'Java', 'Python', 'JavaScript', 'C++', 'Go'}


# 3. remove() vs discard() [Exam ke liye bohot zaroori concept!]
# remove() se delete karna:
my_set.remove("Java") 

# Agar main fir se my_set.remove("Java") likhunga, toh Python ERROR (KeyError) de dega.
# Isliye hum use karte hain discard():
my_set.discard("Java")               # Java pehle hi delete ho chuka hai, par yeh line ERROR nahi degi!
print("After discard:", my_set)


# 4. pop() — Randomly kuch bhi nikalna
# Kyunki set unordered hota hai, pop() kisi bhi ek element ko nikaal dega
deleted_item = my_set.pop()
print(f"Popped item: {deleted_item}")
print("After pop:", my_set)


# 5. clear() — Sab saaf
my_set.clear()
print("After clear:", my_set)        # Output: set() (Khaali set)



# ==============================================================================================
#                                       SET COPY METHOD
# ==============================================================================================
#  Method     | Syntax               | What it does
# ------------|----------------------|----------------------------------------------------------
#  copy       | new_set = set.copy() | Set ki ek alag independent copy banata hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLE ---

# Humare paas ek original users ka set hai
original_users = {"Amit", "Rahul", "Priya"}

# Humne copy() ka use karke ek backup set bana liya
backup_users = original_users.copy()

# Ab agar hum backup wale set mein naye users add ya remove karte hain:
backup_users.add("Rohan")
backup_users.remove("Amit")

# Dono ko print karke check karte hain:
print("Original Users:", original_users) 
# Output: {'Amit', 'Rahul', 'Priya'} -> (Yeh ekdum safe hai, isme koi badlav nahi hua!)

print("Backup Users  :", backup_users)   
# Output: {'Rahul', 'Priya', 'Rohan'} -> (Sirf backup wala change hua)
```

```python
#remove duplicates
l=[1, 2, 3, 4, 3, 2, 4, 5, 7]
print(set(l))

#find common sunjects
sub1={1, 2, 3, 4, 5, 6, 7}
sub2={2, 3, 9, 12, 13, 6}
print(sub1&sub2)

#compare tuples and sets
# ==============================================================================================
#                                   COMPARING TUPLES AND SETS
# ==============================================================================================
#  Feature          | Tuple ()                               | Set {}
# ------------------|----------------------------------------|----------------------------------
#  Syntax           | Parentheses -> (1, 2, 3)               | Curly braces -> {1, 2, 3}
#  Mutability       | IMMUTABLE (Cannot be changed).         | MUTABLE (Can add/remove items).
#  Ordering         | ORDERED (Maintains sequence/index).    | UNORDERED (Random order).
#  Indexing/Slicing | ALLOWED (e.g., tup[0] or tup[1:3]).    | NOT ALLOWED (Throws TypeError).
#  Duplicates       | ALLOWED (Can have duplicate values).   | NOT ALLOWED (Automatically removed).
#  Empty Creation   | empty_tup = ()                         | empty_set = set() (Not {})
# ==============================================================================================

# --- LIVE CODE EXAMPLES TO PROVE THE DIFFERENCES ---

# 1. Duplicate Handling
tup_example = (1, 2, 2, 3)
set_example = {1, 2, 2, 3}

print("Tuple (Keeps duplicates) :", tup_example)  # Output: (1, 2, 2, 3)
print("Set (Removes duplicates):", set_example)  # Output: {1, 2, 3}


# 2. Indexing vs Unordered
print("Tuple Index 0            :", tup_example[0]) # Output: 1
# print(set_example[0]) <-- This will crash! TypeError: 'set' object is not subscriptable


# 3. Changing data (Mutability)
# tup_example[0] = 99 <-- This will crash! TypeError: 'tuple' object does not support item assignment

set_example.add(5)    # Perfectly fine! Sets can be changed.
print("Updated Set :", set_example)  # Output: {1, 2, 3, 5}



#employee tuple and unique cities
```
