```python
#### SET OPERATIONS

#### ==============================================================================================
####                                   THE BIG FOUR SET OPERATIONS
#### ==============================================================================================
####  Operation            | Symbol | Method                 | What it does
#### ----------------------|--------|------------------------|-------------------------------------
####  Union                |   |    | .union()               | Dono sets ke saare elements ko mila deta hai (removes duplicates).
####  Intersection         |   &    | .intersection()        | Sirf wahi elements rakhta hai jo dono sets mein common hain.
####  Difference           |   -    | .difference()          | Pehle set ke wo elements rakhta hai jo dusre set mein nahi hain.
####  Symmetric Difference |   ^    | .symmetric_difference()| Wo elements rakhta hai jo kisi ek set mein hain, par dono mein nahi.
#### ==============================================================================================





#### Quick Code Example to test it:
set_a = {1, 2, 3}
set_b = {3, 4, 5}

#### 1. Union Example
print("Union:", set_a | set_b)                      # Output: {1, 2, 3, 4, 5}

#### 2. Intersection Example
print("Intersection:", set_a.intersection(set_b))    # Output: {3}

# Dono sets mein 'Amit' aur 'Priya' common hain
dance_club = {"Rahul", "Anjali", "Amit", "Priya"}
music_club = {"Amit", "Priya", "Rohan", "Siddharth"}

# Dance club minus Music club (Sirf Pure Dancers bachenge)
only_dance = dance_club - music_club
print("Difference (Dance - Music):", only_dance)
# Output: {'Rahul', 'Anjali'}

# Music club minus Dance club (Sirf Pure Musicians bachenge)
only_music = music_club - dance_club
print("Difference (Music - Dance):", only_music)
# Output: {'Rohan', 'Siddharth'}


# Wo log jo ya toh sirf dance mein hain ya sirf music mein (Common wale gayab)
either_but_not_both = dance_club ^ music_club
print("Symmetric Difference:", either_but_not_both)
# Output: {'Rahul', 'Anjali', 'Rohan', 'Siddharth'}
```

#### SET METHODS

```python
# ==============================================================================================
#                                   IMPORTANT SET METHODS
# ==============================================================================================
#  Method     | Syntax               | What it does
# ------------|----------------------|----------------------------------------------------------
#  add        | set.add(item)        | Set mein ek naya element add karta hai.
#  update     | set.update(iterable) | Ek sath kayi saare elements (list ya set) add karta hai.
#  remove     | set.remove(item)     | Element delete karta hai. Agar element na ho toh ERROR deta hai.
#  discard    | set.discard(item)    | SAFE DELETE! Element na bhi ho toh ERROR NAHI deta.
#  pop        | set.pop()            | Kisi bhi RANDOM element ko nikaal kar delete kar deta hai.
#  clear      | set.clear()          | Set ke saare elements ko uda kar usey khaali (empty) kar deta hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

my_set = {"Python", "Java"}

# 1. add() — Ek item add karna
my_set.add("C++")
print("After add:", my_set)          # Output: {'Python', 'Java', 'C++'} (Order random ho sakta hai)


# 2. update() — Ek sath multiple items lana
my_set.update(["JavaScript", "Go"])
print("After update:", my_set)       # Output: {'Java', 'Python', 'JavaScript', 'C++', 'Go'}


# 3. remove() vs discard() [Exam ke liye bohot zaroori concept!]
# remove() se delete karna:
my_set.remove("Java") 

# Agar main fir se my_set.remove("Java") likhunga, toh Python ERROR (KeyError) de dega.
# Isliye hum use karte hain discard():
my_set.discard("Java")               # Java pehle hi delete ho chuka hai, par yeh line ERROR nahi degi!
print("After discard:", my_set)


# 4. pop() — Randomly kuch bhi nikalna
# Kyunki set unordered hota hai, pop() kisi bhi ek element ko nikaal dega
deleted_item = my_set.pop()
print(f"Popped item: {deleted_item}")
print("After pop:", my_set)


# 5. clear() — Sab saaf
my_set.clear()
print("After clear:", my_set)        # Output: set() (Khaali set)



# ==============================================================================================
#                                       SET COPY METHOD
# ==============================================================================================
#  Method     | Syntax               | What it does
# ------------|----------------------|----------------------------------------------------------
#  copy       | new_set = set.copy() | Set ki ek alag independent copy banata hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLE ---

# Humare paas ek original users ka set hai
original_users = {"Amit", "Rahul", "Priya"}

# Humne copy() ka use karke ek backup set bana liya
backup_users = original_users.copy()

# Ab agar hum backup wale set mein naye users add ya remove karte hain:
backup_users.add("Rohan")
backup_users.remove("Amit")

# Dono ko print karke check karte hain:
print("Original Users:", original_users) 
# Output: {'Amit', 'Rahul', 'Priya'} -> (Yeh ekdum safe hai, isme koi badlav nahi hua!)

print("Backup Users  :", backup_users)   
# Output: {'Rahul', 'Priya', 'Rohan'} -> (Sirf backup wala change hua)
```

```python

# Our base dictionary
car = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}

# 1. Accessing data using a Key
print(car["model"])  # Output: Mustang

# 2. Modifying/Updating a value
car["year"] = 2026   # Changes 1964 to 2026
print(car)

# 3. Adding a completely new Key-Value pair
car["color"] = "Red" 
print(car)           # Output will now include 'color': 'Red'
```

```python
# What happens if a Key doesn't exist?
# If you try to run 

# print(car["price"])
# , Python will crash with a KeyError. To avoid this crash, Python provides a super safe alternative method called .get():


# Safe accessing
print(car.get("price"))  # Output: None (No crash!)

# You can even set a fallback default message if it's missing
print(car.get("price", "Not Available"))  # Output: Not Available
```

```python
# ==============================================================================================
#                                   IMPORTANT DICTIONARY METHODS
# ==============================================================================================
#  Method     | Syntax                 | What it does
# ------------|------------------------|--------------------------------------------------------
#  keys()     | dict.keys()            | Dictionary ki saari KEYS ki ek list (view) deta hai.
#  values()   | dict.values()          | Dictionary ki saari VALUES ki ek list (view) deta hai.
#  items()    | dict.items()           | Saari (Key, Value) ki jodiyan tuples ke roop mein deta hai.
#  get()      | dict.get(key, default) | SAFE ACCESS! Key na ho toh error dene ki jagah default value deta hai.
#  update()   | dict.update(other_dict)| Ek sath kai saari key-value jodiya add ya change karta hai.
#  pop()      | dict.pop(key)          | Kisi specific key ko delete karta hai aur uski value return karta hai.
#  popitem()  | dict.popitem()         | Sabse LAST wali key-value jodi ko delete karke return karta hai.
#  clear()    | dict.clear()           | Dictionary ko poori tarah se khaali ({}) kar deta hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

student = {
    "name": "Amit",
    "age": 20,
    "course": "Python"
}

# 1. keys(), values(), aur items() -> Loop chalane ke liye bohot kaam aate hain
print("Saari Keys  :", student.keys())    # Output: dict_keys(['name', 'age', 'course'])
print("Saare Values:", student.values())  # Output: dict_values(['Amit', 20, 'Python'])
print("Saare Items :", student.items())   # Output: dict_items([('name', 'Amit'), ('age', 20), ('course', 'Python')])


# 2. update() -> Ek sath naya data jodne ya badalne ke liye
student.update({"age": 21, "city": "Delhi"}) # age change ho gayi, city nayi add ho gayi
print("\nAfter update():", student)


# 3. pop() vs popitem() -> Elements delete karne ke liye
# pop() mein key batani padti hai:
deleted_value = student.pop("course")
print(f"\npop('course') ne '{deleted_value}' delete kiya. Dict:", student)

# popitem() bina kisi argument ke last element udata hai:
deleted_pair = student.popitem()
print(f"popitem() ne {deleted_pair} ko uda diya. Dict:", student)


# 4. clear() -> Sab saaf
student.clear()
print("\nAfter clear():", student)  # Output: {} (Khaali dictionary)


user_scores = {"Rahul": 90, "Priya": 95}

for key, value in user_scores.items():
    print(f"Student: {key}, Score: {value}")
```

```python
# ==============================================================================================
#                               ADDITIONAL DICTIONARY METHODS
# ==============================================================================================
#  Method      | Syntax                        | What it does
# -------------|-------------------------------|------------------------------------------------
#  copy        | new_dict = dict.copy()        | Dictionary ki ek alag independent copy banata hai.
#  fromkeys    | dict.fromkeys(keys, value)    | Ek list/sequence se nayi dictionary banata hai jisme sabki value SAME hoti hai.
#  setdefault  | dict.setdefault(key, default) | Agar key hai, toh uski value deta hai. Agar NAHI hai, toh naya key-value insert kar deta hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

# 1. copy() — Independent duplicate banana
original = {"a": 1, "b": 2}
duplicate = original.copy()

duplicate["c"] = 3
print("Original after modifying copy:", original)   # Output: {'a': 1, 'b': 2} -> (Original ekdum safe hai!)
print("Duplicate Set                 :", duplicate)  # Output: {'a': 1, 'b': 2, 'c': 3}


# 2. fromkeys() — Default ya khali templates banane ke liye
# Maan lo hume 3 students ke liye ek dict banani hai aur sabki initial score status 'Pending' rakhna hai
students_list = ["Rahul", "Priya", "Amit"]
default_value = "Pending"

new_status_dict = dict.fromkeys(students_list, default_value)
print("\nAfter fromkeys():", new_status_dict)
# Output: {'Rahul': 'Pending', 'Priya': 'Pending', 'Amit': 'Pending'}


# 3. setdefault() — [Sabse bada head-spinner method!]
# Yeh do kaam ek sath karta hai:
# Case A: Agar key PEHLE SE HAI, toh yeh uski purani value hi return karega (kuch badlega nahi).
profile = {"name": "Anjali", "role": "Admin"}
existing_val = profile.setdefault("role", "Guest") 

print("\nCase A (Key exists) -> Returned Val:", existing_val) # Output: Admin
print("Profile stays same                :", profile)      # Output: {'name': 'Anjali', 'role': 'Admin'}

# Case B: Agar key MISSING HAI, toh yeh chupchaap naye key aur default value ko insert kar dega.
missing_val = profile.setdefault("city", "Delhi")

print("Case B (Key missing) -> Returned Val:", missing_val)  # Output: Delhi
print("Profile updated with new data       :", profile)      # Output: {'name': 'Anjali', 'role': 'Admin', 'city': 'Delhi'}
```

```python
# ==============================================================================================
#                               BUILT-IN FUNCTIONS WITH DICTIONARIES
# ==============================================================================================
#  Function | Syntax            | What it does
# ----------|-------------------|---------------------------------------------------------------
#  len()    | len(dict)         | Dictionary mein total kitni key-value jodiya (pairs) hain, batata hai.
#  sorted() | sorted(dict)      | Dictionary ki KEYS ko sort karke ek nayi list return karta hai.
#  all()    | all(dict)         | Agar SAARI keys True (non-zero/non-empty) hain, toh True deta hai.
#  any()    | any(dict)         | Agar KOI BHI EK key True hai, toh True deta hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

inventory = {
    "Laptop": 10,
    "Mouse": 50,
    "Keyboard": 30
}

# 1. len() — Total count nikalna
print("Total Items in Inventory:", len(inventory))  
# Output: 3 (Kyunki 3 pairs hain)


# 2. sorted() — Keys ko alphabetic order mein lagana
sorted_keys = sorted(inventory)
print("Sorted Keys List        :", sorted_keys)    
# Output: ['Keyboard', 'Laptop', 'Mouse']

# Pro Tip: Agar values ke hisab se sort karna ho (Exam trick):
sorted_by_values = sorted(inventory, key=inventory.get)
print("Sorted by Quantity      :", sorted_by_values) 
# Output: ['Laptop', 'Keyboard', 'Mouse'] (10, 30, 50 ke order mein)


# 3. all() vs any() — Boolean check (Yeh hamesha KEYS par check karta hai, values par nahi!)
# Case A: Saari keys valid strings hain (True)
print("all(inventory) status   :", all(inventory))  # Output: True

# Case B: Agar koi ek key 0 ya Empty string "" ho jaye
corrupted_dict = {
    "ValidKey": "Yes",
    0: "This key is zero/False"
}

print("any(corrupted_dict)     :", any(corrupted_dict)) # Output: True (Kyunki ek key True hai)
print("all(corrupted_dict)     :", all(corrupted_dict)) # Output: False (Kyunki 0 key False maani jaati hai)
```

```python
# ==============================================================================================
#                                   NESTED DICTIONARY STRUCTURE
# ==============================================================================================
#  Concept | Syntax                                       | What it does
# ---------|----------------------------------------------|-------------------------------------
#  Nested  | main_dict = { key1: {sub_key: value}, ... }  | Ek dictionary ki VALUE khud ek naya
#          |                                              | dictionary hoti hai.
# ==============================================================================================

# --- LIVE CODE EXAMPLE ---

# Maan lo hamare paas ek school class ka data hai
classroom = {
    "student1": {
        "name": "Amit",
        "age": 20,
        "marks": {"Maths": 90, "Python": 95}  # Ek aur nesting!
    },
    "student2": {
        "name": "Priya",
        "age": 21,
        "marks": {"Maths": 85, "Python": 98}
    }
}

# ---------------------------------------------------------
# 1. ACCESSING DATA (Double Square Brackets '[][]')
# ---------------------------------------------------------
# Pehle main key pakdo, fir andar wali key
print("Student 1 ka Naam:", classroom["student1"]["name"])  
# Output: Amit

# Agar teesre level tak jana ho:
print("Student 2 ke Python Marks:", classroom["student2"]["marks"]["Python"])  
# Output: 98


# ---------------------------------------------------------
# 2. UPDATING / MODIFYING DATA
# ---------------------------------------------------------
# Amit ki age 20 se 22 karni hai
classroom["student1"]["age"] = 22
print("\nAmit ki updated age:", classroom["student1"]["age"])  # Output: 22


# ---------------------------------------------------------
# 3. ADDING NEW DATA INSIDE NESTED DICT
# ---------------------------------------------------------
# Priya ke data mein ek naya phone number jodhna hai
classroom["student2"]["phone"] = "9876543210"
print("\nPriya ka naya data:", classroom["student2"])
# Output: {'name': 'Priya', 'age': 21, 'marks': {...}, 'phone': '9876543210'}


# ---------------------------------------------------------
# 4. LOOPING THROUGH A NESTED DICTIONARY
# ---------------------------------------------------------
print("\n--- Printing All Students Info ---")
for student_id, info in classroom.items():
    print(f"ID: {student_id}")
    print(f"Name: {info['name']}, Age: {info['age']}")
```

```python
# ==============================================================================================
#                               DICTIONARY COMPREHENSION SYNTAX
# ==============================================================================================
#  Syntax (Basic) : new_dict = {key_expression: value_expression for item in iterable}
#  Syntax (With If): new_dict = {key_expr: value_expr for item in iterable if condition}
# ==============================================================================================

# --- LIVE CODE EXAMPLES ---

# ---------------------------------------------------------
# Example 1: Basic Comprehension (Numbers aur unke Squares)
# ---------------------------------------------------------
# Agar hume 1 se 5 tak ke numbers ka square store karna ho:

# Aam Tarika (Normal Loop):
squares_normal = {}
for x in range(1, 6):
    squares_normal[x] = x**2

# Shortcut Tarika (Dictionary Comprehension):
squares_fast = {x: x**2 for x in range(1, 6)}
print("Squares Dict:", squares_fast)
# Output: {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}


# ---------------------------------------------------------
# Example 2: Comprehension with 'if' Condition (Filtering)
# ---------------------------------------------------------
# Maan lo hamare paas student ke marks hain, aur hume sirf PASS hone wale (>40) chahiye:
all_marks = {"Rahul": 35, "Priya": 85, "Amit": 42, "Anjali": 90}

# Sirf unko select karo jinke marks >= 40 hain
passed_students = {name: score for name, score in all_marks.items() if score >= 40}
print("\nPassed Students Only:", passed_students)
# Output: {'Priya': 85, 'Amit': 42, 'Anjali': 90}  (Rahul delete ho gaya!)


# ---------------------------------------------------------
# Example 3: Comprehension with 'if-else' (Transforming)
# ---------------------------------------------------------
# Agar hume marks ke hisab se 'Pass' ya 'Fail' status store karna ho:
status_dict = {name: ("Pass" if score >= 40 else "Fail") for name, score in all_marks.items()}
print("\nStudents Status   :", status_dict)
# Output: {'Rahul': 'Fail', 'Priya': 'Pass', 'Amit': 'Pass', 'Anjali': 'Pass'}
```

```python
# 1. Create a student dict
student = {
    "roll_no": 101,
    "name": "Amit Kumar",
    "course": "BTech"
}
print("1. Initial Student Dict:", student)


# 2. Add a new key
student["city"] = "Delhi"
print("2. After Adding Key   :", student)


# 3. Update an existing value
student["course"] = "Data Science"
print("3. After Updating Value:", student)


# 4. Delete a key using pop()
deleted_val = student.pop("roll_no")
print(f"4. Popped Value        : {deleted_val}")
print("   Dict After pop()    :", student)


# 5. Print all keys
print("5. All Keys            :", list(student.keys()))


# 6. Print all values
print("6. All Values          :", list(student.values()))


# 7. Traverse using items
print("\n7. Traversing using items():")
for key, value in student.items():
    print(f"   Key: {key:8} | Value: {value}")


# 8. Create a nested dict
classroom = {
    "sec_A": {"total": 40, "monitor": "Rahul"},
    "sec_B": {"total": 42, "monitor": "Priya"}
}
print("\n8. Nested Dictionary   :", classroom)


# 9. Create a dict comprehension for cubes
# Generates numbers 1 to 5 mapping to their cubes
cubes_dict = {num: num**3 for num in range(1, 6)}
print("9. Cubes Dict Comp     :", cubes_dict)


# 10. Create freq of chars in a string using dict
sample_string = "programming"
frequency = {}

for char in sample_string:
    # get(char, 0) checks if character is already there, else returns default 0
    frequency[char] = frequency.get(char, 0) + 1

print("10. Char Frequency     :", frequency)
```
