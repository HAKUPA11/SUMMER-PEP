```python
import pandas as pd
```

```python
std=pd.read_csv("student_performance_project.csv")
std.head(10)
```

Step 2 – First Look at Data Quality

```python
# 1. Display first 10 rows
print("First 10 Rows:")
print(std.head(10))

# 2. Shape of the dataset
print("\nShape of Dataset:")
print(std.shape)

# 3. Column names
print("\nColumn Names:")
print(std.columns)

# 4. Information about the dataset
print("\nDataset Info:")
print(std.info())

# 5. Statistical Summary
print("\nStatistical Summary:")
print(std.describe())

# 6. Check missing values
print("\nMissing Values:")
print(std.isnull().sum())

# 7. Count duplicate rows
print("\nDuplicate Rows:")
print(std.duplicated().sum())
```

Step 3.1 – Display Rows with Missing Values

```python
print("Rows with Missing Values:")
print(std[std.isnull().any(axis=1)])
```

Step 3.2 – Count Missing Values Again

```python
print(std.isnull().sum())
```

Step 3.3 – Fill Missing Values

```python
#version 1
std["Math"] = std["Math"].fillna(std["Math"].mean())
std["Science"] = std["Science"].fillna(std["Science"].mean())
std["English"] = std["English"].fillna(std["English"].mean())

#version 2mei we use inplace=True
std.fillna({
    "Math": std["Math"].mean(),
    "Science": std["Science"].mean(),
    "English": std["English"].mean()
}, inplace=True)
```

```python
#then we can verify also
print(std.isnull().sum())
```

Step 4 – Removing Duplicate Rows

```python
print("Number of Duplicate Rows:", std.duplicated().sum())
```

4.2 Display the duplicate row(s)

```python
print(std[std.duplicated()])
```

4.3 Remove duplicate rows

```python
std.drop_duplicates(inplace=True)
```

4.4 Verify

```python
print("Number of Duplicate Rows:", std.duplicated().sum())
print("New Shape:", std.shape)
```

Step 5 – Filtering Students

```python
high_math = std[std["Math"] > 80]
print(high_math)
```

Students from Delhi

```python
delhi_students = std[std["City"] == "Delhi"]
print(delhi_students)
```

Female students

```python
female_students = std[std["Gender"] == "F"]
print(female_students)
```

Students scoring more than 75 in both Math and Science

```python
top_students = std[(std["Math"] > 75) & (std["Science"] > 75)]
print(top_students)
```

Students from Mumbai who scored more than 60 in English

```python
mumbai_students = std[(std["City"] == "Mumbai") & (std["English"] > 60)]
print(mumbai_students)
```

Step 6 – Subject-wise Analysis

6.1 Average Marks in Each Subject

```python
print("Average Marks")
print("Math:", std["Math"].mean())
print("Science:", std["Science"].mean())
print("English:", std["English"].mean())
```

6.2 Highest Marks in Each Subject

```python
print("Highest Marks")
print("Math:", std["Math"].max())
print("Science:", std["Science"].max())
print("English:", std["English"].max())
```

6.3 Lowest Marks in Each Subject

```python
print("Lowest Marks")
print("Math:", std["Math"].min())
print("Science:", std["Science"].min())
print("English:", std["English"].min())
```

6.4 Total Marks of Each Student

```python
std["Total"] = std["Math"] + std["Science"] + std["English"]

print(std.head())
```

6.5 Average Marks of Each Student

```python
std["Average"] = std["Total"] / 3

print(std.head())
```

6.6 Find the Top Performer

```python
top_student = std.loc[std["Total"].idxmax()]

print(top_student)
```

6.7 Find the Lowest Performer

```python
low_student = std.loc[std["Total"].idxmin()]

print(low_student)
```

6.8 Subject-wise Statistics Together

```python
print(std[["Math", "Science", "English"]].describe())
```

Step 7 – Total, Average and Ranking

Calculate Total Marks

```python
std["Total"] = std["Math"] + std["Science"] + std["English"]
```

Calculate Average Marks

```python
std["Average"] = std["Total"] / 3
```

Rank Students

Highest marks should get Rank 1.

```python
std["Rank"] = std["Total"].rank(method="dense", ascending=False)

print(std[["Name","Total","Average","Rank"]])
```

Step 8 – Pass / Fail Analysis

Suppose passing marks are 35 in every subject.

```python
std["Result"] = std.apply(
    lambda x: "Pass"
    if x["Math"] >= 35 and x["Science"] >= 35 and x["English"] >= 35
    else "Fail",
    axis=1
)

print(std[["Name","Result"]])
```

Count Pass and Fail Students

```python
print(std["Result"].value_counts())
```

Step 9 – Gender-wise and City-wise Analysis

Gender-wise Average

```python
print(std.groupby("Gender")[["Math","Science","English","Average"]].mean())
```

City-wise Average

```python
print(std.groupby("City")[["Math","Science","English","Average"]].mean())
```

Number of Students in Each City

```python
print(std["City"].value_counts())
```

Number of Male and Female Students

```python
print(std["Gender"].value_counts())
```

Step 10 – Combined GroupBy Report

This combines City and Gender.

```python
report = std.groupby(["City","Gender"])[["Math","Science","English","Average"]].mean()

print(report)
```

Step 11 – Topper List

top 5 students

```python
top5 = std.sort_values(by="Total", ascending=False).head(5)

print(top5[["Name","Total","Average","Rank"]])
```

Or Top 10

```python
top10 = std.nlargest(10, "Total")

print(top10[["Name","Total"]])
```

Step 12 – Exporting Report

```python
std.to_csv("student_performance_report.csv", index=False)

print("Report Exported Successfully!")
```
