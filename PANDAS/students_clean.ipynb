```python
# merge, concatenate, join
od=pd.read_csv("orders.csv")
cu=pd.read_csv("customers.csv")
```

```python
od.head()
```

```python
cu.head()
```

1. Merge (pd.merge)
When to use: When you want to combine two tables based on a shared column/key (in this case, matching the customer identifier, such as CustID). This is a relational join.

```python
# Assuming the customer ID column in 'cu' is named 'CustID'
merged_df = pd.merge(od, cu, on="CustID", how="inner")
print(merged_df.head())
```

2. Join (df.join)
When to use: When you want to combine two DataFrames based on their Indices rather than specific columns.

If you want to use .join() using your customer identifier, you first need to set CustID as the index for both tables:

```python
print(od.set_index('CustID').join(cu.set_index('CustID'), how="inner"))
```
