```python
#reverse a string
s = input("word bolo: ")
i, j=0, len(str)-1

l=list(s)

print(l)

while(i<j):
    l[i], l[j] = l[j], l[i]
    i+=1
    j-=1

rev_s="".join(l)
print(rev_s)

#-------------
```

```python
#check palindrome:
s=str(input("shabd likhe: "))

i, j=0, len(s)-1
l=list(s)

print(l)

while(i<j):
    if(l[i]==l[j]):
        i+=1
        j-=1

```
