1) Handle Division by Zero
Catches the ZeroDivisionError when attempting to divide any number by zero.

```python
try:
    num1 = 10
    num2 = 0
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Division by zero is mathematically impossible!")
```

2) Handle Invalid Age Input
Catches ValueError if a user types text instead of a numerical integer for their age.

```python
try:
    age = int(input("Enter your age: "))
    print(f"Your age is registered as: {age}")
except ValueError:
    print("Error: Invalid entry! Age must be a whole number.")
    
```

3) Handle List Index Error
Catches IndexError when trying to fetch an element beyond the size boundaries of a list.

```python
fruits = ["Apple", "Banana", "Cherry"]

try:
    # Indexing starts at 0, max index is 2. Index 5 does not exist.
    print(fruits[5])
except IndexError:
    print("Error: The requested index is out of bounds for this list.")
```

4) Handle Dictionary Key Error
Catches KeyError when attempting to access a key that is missing from a dictionary.

```python
student = {"name": "Amit", "course": "Python"}

try:
    print(student["roll_no"])
except KeyError:
    print("Error: The requested data key does not exist in the record.")
```

5) Handle File Not Found
Catches FileNotFoundError when a script attempts to read a file that isn't present in the system directory.

```python
try:
    with open("ghost_file.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("Error: The target file could not be found in the directory.")
    
```

6) Calculator Using Try-Except
A bulletproof mini-calculator that catches both mathematical errors and faulty input values safely.

```python
try:
    a = float(input("Enter first number: "))
    op = input("Enter operator (+, -, *, /): ")
    b = float(input("Enter second number: "))

    if op == '+': result = a + b
    elif op == '-': result = a - b
    elif op == '*': result = a * b
    elif op == '/': result = a / b
    else: result = "Invalid Operator!"

    print(f"Result: {result}")
except ZeroDivisionError:
    print("Math Error: Cannot divide by zero.")
except ValueError:
    print("Input Error: Please enter valid numeric characters only.")
```

7) Validate Salary (>0)
Uses the raise keyword to manually trigger a generic ValueError exception if the input value falls below a logical threshold.

```python
try:
    salary = float(input("Enter salary amount: "))
    if salary <= 0:
        raise ValueError("Salary must be a positive number greater than 0.")
    print(f"Salary approved: ${salary}")
except ValueError as error:
    print(f"Validation Error: {error}")
```

8) Validate Password Length
Enforces an arbitrary minimum rule limit (e.g., at least 6 characters long) using defensive error raising blocks.

```python
try:
    password = input("Create a password: ")
    if len(password) < 6:
        raise ValueError("Password is too short! It must be at least 6 characters long.")
    print("Password structure accepted.")
except ValueError as error:
    print(f"Security Alert: {error}")
```

9) Employee Bonus Calculator
Safely calculates performance incentives based on structural input validation tracking.

```python
try:
    base_salary = float(input("Enter Base Salary: "))
    rating = int(input("Enter Performance Rating (1 to 5): "))
    
    if rating < 1 or rating > 5:
        raise ValueError("Rating scale must be between 1 and 5 only.")
        
    # Bonus tier math
    bonus_percentage = rating * 5  # 5% to 25% max
    bonus_amount = base_salary * (bonus_percentage / 100)
    
    print(f"Calculated Bonus: ${bonus_amount:.2f} (Total Payout: ${base_salary + bonus_amount:.2f})")
except ValueError as error:
    print(f"Processing Error: {error}")
```

10) Shopping Bill with Invalid Quantity
Safeguards retail cart checkout calculations against fractional, negative, or alphabetical inputs.

```python
price_per_item = 49.99

try:
    qty_input = input("Enter total items to purchase: ")
    quantity = int(qty_input)  # Catches non-integers (like 2.5 or 'two')
    
    if quantity <= 0:
        raise ValueError("Quantity must be at least 1 unit.")
        
    total_bill = price_per_item * quantity
    print(f"Checkout Complete! Your total bill is: ${total_bill:.2f}")
except ValueError as error:
    # Handles both invalid conversion string errors and custom raised condition blocks
    if "invalid literal" in str(error):
        print("Cart Error: Quantity must be a clean whole number.")
    else:
        print(f"Cart Error: {error}")
```
