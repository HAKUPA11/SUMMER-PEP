# MATHEMATICAL FUNCTIONS

```python
#check the directory we can use it with any packages
import numpy as np
dir(np)
```

```python
a=np.array([1, 2, 3, 4, 5])
print(np.sum(a))
print(np.max(a))
print(np.min(a))
print(np.(a))
print(np.(a))
```

## 1. Arithmetic Functions

```python
a = np.array([1, 2, 3, 4])

print(np.add(a, 5))
print(np.subtract(a, 1))
print(np.multiply(a, 2))
print(np.divide(a, 2))
print(np.power(a, 2))
print(np.mod(a, 2))
```

## 2. Exponential and Logarithmic Functions

```python
a = np.array([1, 2, 3])

print(np.exp(a))      # e^x
print(np.log(a))      # Natural log
print(np.log10(a))    # Base-10 log
print(np.log2(a))     # Base-2 log
```

## 3. Square Root and Powers

```python
a = np.array([1, 4, 9, 16])

print(np.sqrt(a))
print(np.square(a))
print(np.cbrt(a))
```

## 4. Trigonometric Functions

```python
angles = np.array([0, np.pi/2, np.pi])

print(np.sin(angles))
print(np.cos(angles))
print(np.tan(angles))

print(np.arcsin([0, 1]))
print(np.arccos([1, 0]))
print(np.arctan([0, 1]))
```

## 5. Rounding Functions

```python
a = np.array([2.3, 2.7, -2.3, -2.7])

print(np.floor(a))
print(np.ceil(a))
print(np.round(a))
print(np.trunc(a))
```

## 6. Absolute Value

```python
a = np.array([-3, -2, 5])

print(np.abs(a))
```

## 7. Statistical Functions

```python
a = np.array([2, 4, 6, 8, 10])

print(np.sum(a))
print(np.mean(a))
print(np.median(a))
print(np.min(a))
print(np.max(a))
print(np.std(a))
print(np.var(a))
```

## 8. Cumulative Functions

```python
a = np.array([1, 2, 3, 4])

print(np.cumsum(a))
print(np.cumprod(a))
```

## 9. Comparison Functions

```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print(np.maximum(a, b))
print(np.minimum(a, b))
```

## 10. Sign Functions

```python
a = np.array([-5, 0, 8])

print(np.sign(a))
```

## 11. Clipping Values

```python
a = np.array([1, 5, 10, 15])

print(np.clip(a, 3, 12))
```

## 12. Linear Algebra Functions

```python
A = np.array([[1, 2], [3, 4]])

print(np.linalg.det(A))
print(np.linalg.inv(A))
print(np.linalg.eig(A))
```

## 13. Common Constants

```python
print(np.pi)
print(np.e)
print(np.inf)
print(np.nan)
```
