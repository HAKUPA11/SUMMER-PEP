```python
calc=input("enter your calculation: ")
a=input("enter your first number: ")
b=input("enter your second number: ")
if calc=="+":
    print("the addition of", a,"and", b,":",float(a)+float(b))
elif calc=="-":
    print(float(a)-float(b)) 
elif calc=="*":
    print(float(a)*float(b))    
elif calc=="/":
    print(float(a)/float(b))
elif calc=="%":
    print(float(a)%float(b))
elif calc=="**":
    print(float(a)**float(b))
elif calc=="//":
    print(float(a)//float(b))
else:    print("invalid operator")
```
