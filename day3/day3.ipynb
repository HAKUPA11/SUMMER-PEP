```python
#check if the input is palindrome or not
s = input()
i = 0
j = len(s) - 1

is_palin = True

while i < j:
    if s[i] != s[j]:
        is_palin = False
        break

    i += 1
    j -= 1

if is_palin:
    print(s)
    print("Palindrome")
else:
    print(s)
    print("Not Palindrome")
```

```python
# revrese a number
num = int(input("Enter a number: "))
og=num
rev = 0

while num > 0:
    dgt = num % 10
    rev = rev * 10 + dgt
    num = num // 10

print("Original number=", og)
print("Reversed number =", rev)
```

```python
# primt number

num = int(input("Enter a number: "))
og=num
if num <= 1:
    print("Not Prime")
else:
    is_prm = True

    for i in range(2, num):
        if num % i == 0:
            is_prm= False
            break

    if is_prm:
        print("the number was",og)
        print("Prime")
    else:
        print("the number was",og)
        print("Not Prime")
```

```python
#fibonacci series
n = int(input("Enter number of terms: "))

a = 0
b = 1
cnt = 0

while cnt < n:
    print(a, end=" ")
    c = a + b
    a = b
    b = c
    cnt += 1
```
