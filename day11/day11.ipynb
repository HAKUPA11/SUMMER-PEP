```python
netflix_shows=['obsession','50 shades of grey','365 days','the girl next door','cruel intention','darker grey']
print("--- Netflix Dashboard ---")
print("First show:", netflix_shows[0])
print("Last show:", netflix_shows[-1])
print("Top 3:", netflix_shows[:3])
print("Last 3:", netflix_shows[-3:])
print("Alternate shows:", netflix_shows[::2])
print("Reverse list:", netflix_shows[::-1])
```

```python
# Q2: Hospital Patient Monitoring - Traversal
heart_rates = [72, 105, 98, 112, 85, 120, 95]
high_risk = [hr for hr in heart_rates if hr > 100]
print("\n--- Hospital Monitoring ---")
print("Patients with HR > 100:", high_risk)
print("Count of high risk patients:", len(high_risk))
```

```python
# Q3: Company Payroll Analytics - Built-in Functions
salaries = [45000, 60000, 32000, 85000, 52000]
print("\n--- Payroll Analytics ---")
print("Highest Salary:", max(salaries))
print("Lowest Salary:", min(salaries))
print("Total Salary:", sum(salaries))
print("Average Salary:", sum(salaries) / len(salaries))
```

```python
# Q4: Employee Access Portal - Membership
authorized_ids = {101, 105, 202, 305}
emp_id = int(input("\nEnter Employee ID: "))
if emp_id in authorized_ids:
    print("Access Granted")
else:
    print("Access Denied")
```

```python
# Q5: Debugging Challenge - Aliasing
# Aliasing happens when two variables point to the same object in memory.
a = [1, 2, 3]
b = a  # Aliasing occurs here
b.append(4)
print("\n--- Debugging Aliasing ---")
print(f"Original 'a' after modifying 'b': {a}")
```
