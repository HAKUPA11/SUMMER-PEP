## MODULES

```python
import math
import random
import datetime
import os
import sys
```

```python
# 1. Calculate powers using math.pow()
base = 5
exponent = 3
power = math.pow(base, exponent)
print(f"{base}^{exponent} = {power}")

# 2. Generate five random numbers using randint()
print("\nFive Random Numbers:")
for i in range(5):
    print(random.randint(1, 100))

# 3. Display today's date using datetime
today = datetime.date.today()
print("\nToday's Date:", today)

# 4. Print the current working directory
print("\nCurrent Working Directory:")
print(os.getcwd())

# 5. Display the Python version
print("\nPython Version:")
print(sys.version)
```
