```python
#Q2

def c(x, y, o):
    if o == '+': return x+y
    if o == '-': return x-y
    if o == '*': return x*y
    if o == 'sq': return x ** 2    
    if o == 'sqrt': return x ** 0.5 
    if o == 'cb': return x ** 3
    if o == '/': return x/y if y != 0 else "Err"
    return "Err"

x=int(input("enter the first number: "))
y=int(input("enter the second number: "))
o=str(input("enter the operation you want to perform: "))

c(x, y, o)
```

```python
def add(a, b):
    return a + b

print(add(10, 20))


def greet(name, city="Delhi"):
    print("Name:", name)
    print("City:", city)

greet("draxi")


def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="hakupa")

def largest(*numbers):
    print("Largest Number:", max(numbers))

largest(10, 50, 30, 90, 40)


def details(**student):
    print(student)

details(name="Aadi", age=20, course="B.Tech")
```
