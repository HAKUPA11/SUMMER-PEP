### STEP 1-- DATASET CREATION

```python
import numpy as np
```

```python
stds = np.array([
    [85, 90, 78, 92, 88],
    [70, 68, 75, 72, 74],
    [95, 98, 100, 96, 99],
    [60, 65, 58, 62, 64],
    [80, 82, 84, 79, 81]
])

print("the students and their marks as follows: \n", stds)
```

### STEP 2--AVERAGE MARKS PER STUDENT

```python
avg=np.mean(stds, axis=1)
print("average per student: \n", avg)
```

### STEP 3-- HIGHEST AND LOWEST MARKS

```python
mini=np.min(stds, axis=1)
print("the minimum marks per student are as follows: \n", mini)

maxi=np.max(stds, axis=1)
print("the maximum marks per student are as follows: \n", maxi)
```

### STEP 4 --STANDARD DEVIATION PER STUDENT

```python
std=np.std(stds, axis=1)

print("the standard deviation per student are as follows: \n", std)
```

### STEP 5 -- RANKING STUDENT

```python
total=sum(stds)
print(total)

print(np.sort(total))
```

### STEP 6 -- TOP STUDENTS (CLASS AVERAGE)

```python
clsavg=np.mean(stds)
print(clsavg)

print("students above class average are \n", np.where(total > clsavg))
```

### STEP 7 -- SUBJECT WISE ANALYSIS

```python
print("subject wise analysis: -", np.mean(stds, axis=0))

print("highest marks in each subject: \n", np.max(stds, axis=0))
print("lowest marks in each subject: \n", np.min(stds, axis=0))
```

### STEP 8 PERCENTAGE CALC

```python
per = (total / 500) * 100

print("Percentage")
print(per)
```
