```python
with open("intro.txt") as f:
    l=f.read().splitlines()

print(l)
f.close()
```

```python
print("2. Reading file contents:")
try:
    with open("intro.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("Error: The file does not exist yet.")
```
